# Analiza i Klasterovanje PBMC Podataka

## Uvod

Periferne molekularne krvne ćelije (PBMCs) su ključne komponente imunog sistema i uključuju različite tipove ćelija kao što su limfociti (B ćelije, T ćelije, NK ćelije), monociti i dendritske ćelije. Ove ćelije igraju vitalnu ulogu u imunološkim odgovorima i koriste se u istraživanju raznih oblasti biomedicine, uključujući infektivne bolesti, imunologiju (uključujući autoimune poremećaje), malignitete, transplatacionu imunologiju, vakcine i skrining.

U ovom projektu analiziraćemo skupove podataka dobijene iz različitih tipova PBMC ćelija. Svaka od dve ulazne datoteke sadrži podatke o ekspresiji 10800 gena, a skup gena je identičan i uređen po istom redosledu u obe datoteke. Nazivi gena imaju prefiks hg38, što znači da su podaci vezani za verziju 38 humanog genoma.

Cilj ovog zadatka je da izvršimo klasterovanje podataka sa najmanje tri različita algoritma za svaki pojedinačni skup podataka i utvrdimo da li dobijeni klasteri odgovaraju poznatim tipovima ćelija. Takođe ćemo primeniti formirane modele klasterovanja na drugu (preostalu) datoteku i ponoviti analizu kvaliteta. Metode koje ćemo koristiti uključuju:

- **UMAP (Uniform Manifold Approximation and Projection)** za redukciju dimenzija.
- **K-means** klasterovanje.
- **Hijerarhijsko** klasterovanje.
- **DBSCAN** klasterovanje.

Pored toga, koristićemo različite evaluacione metrike kao što su Silhouette skor i vizualizacije kako bismo procenili kvalitet klastera.

Krenimo sa implementacijom.

## UMAP Redukcija Dimenzija

UMAP (Uniform Manifold Approximation and Projection) je algoritam za redukciju dimenzija koji se koristi za smanjenje složenih visokodimenzionalnih podataka u niže dimenzionalne prostore, zadržavajući što je moguće više strukturalnih informacija. U ovom slučaju, UMAP se koristi za smanjenje podataka pre primene klasterovanja kako bi se poboljšala efikasnost i preciznost klasterovanja.



```python
X1 = data1.drop(['src_file', 'row_id', 'cells_type'], axis=1)  
X2 = data2.drop(['src_file', 'row_id', 'cells_type'], axis=1)  

# Kreiranje UMAP reduktora bez random_state za maksimalnu paralelizaciju
umap_reducer = UMAP(n_components=2, n_jobs=-1, min_dist=0.1)
# Koristi umap_reducer na podacima
X1_umap = umap_reducer.fit_transform(X1)
X2_umap = umap_reducer.fit_transform(X2)
```

### Korišćenje UMAP rezultata za K-Means
Nakon što su podaci redukovani pomoću UMAP-a, možemo ih koristiti kao ulaz za K-Means algoritam kako bismo identifikovali klastere u 2D prostoru. Ovo često dovodi do boljih i brže postignutih rezultata u poređenju sa klasterovanjem na originalnim visokodimenzionalnim podacima.

Koristimo K-Means algoritam za klasterovanje kako bi istražili optimalan broj klastera na redukovanim podacima dobijenim primenom UMAP-a. Evaluacija kvaliteta klastera se vrši pomoću Silhouette skora, a rezultati su prikazani na grafikonu kako bi se vizualno identifikovao broj klastera koji daje najbolji rezultat.

#### Dataset1


```python
# Lista za čuvanje vrednosti Silhouette Score
silhouette_scores = []

# Iteracija kroz broj klastera od 2 do 10
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X1_umap)
    silhouette_avg = silhouette_score(X1_umap, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Prikaz rezultata na grafikonu
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()
```

Klaster 2: Dominiraju BC (B ćelije), sa nekoliko DC i TC ćelija.


Preklapanje tipova ćelija u klasterima: Očigledno je da postoji preklapanje između različitih tipova ćelija unutar klastera, posebno za TC (T ćelije) koje se pojavljuju u više klastera.
Dominantni tipovi ćelija po klasterima: Neki klasteri pokazuju jasnu dominaciju određenih tipova ćelija, kao što su klasteri sa većinskim TC ili MC ćelijama. Ovo može ukazivati na to da klasterovanje uspešno identifikuje glavne grupe, ali i dalje postoji preklapanje.

```python
# Mogući opsezi za eps i min_samples
eps_values = np.arange(0.0, 2.0, 0.1)  # Može se fino podešavati
min_samples_values = range(3, 10)  # Broj minimalnih uzoraka u okolini
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN

# Rezultati za svaku kombinaciju
results = []

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X1_reduced)
        if len(np.unique(labels)) > 1:  # Ako postoji više od jednog klastera (ignorišemo slučaj sa samo šumom)
            silhouette = silhouette_score(X1_reduced, labels)
            results.append((eps, min_samples, silhouette, labels))

# Sortiramo rezultate prema Silhouette skoru
results.sort(key=lambda x: x[2], reverse=True)

# Prikazujemo najbolji rezultat
best_eps, best_min_samples, best_silhouette, best_labels = results[0]
print(f"Best eps: {best_eps}, Best min_samples: {best_min_samples}, Best Silhouette Score: {best_silhouette}")

# Vizualizacija najboljeg klasterovanja
plt.figure(figsize=(10, 8))
plt.scatter(X1_reduced[:, 0], X1_reduced[:, 1], c=best_labels, cmap='viridis', s=5)
plt.title(f'DBSCAN Clustering with eps={best_eps}, min_samples={best_min_samples}')
plt.colorbar()
plt.show()
```
